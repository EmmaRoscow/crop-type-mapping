```python
import torch
import torch.nn as nn
```

```python
from recurrent_norm import RecurrentNorm2d
```

```python

timesteps = 20
rn = RecurrentNorm2d(128, timesteps)


# timesteps, batch size, channels, h, w
input_seq = torch.randn(timesteps, 5, 128, 64, 64)

print("THESE VALUES SHOULD BE THE SAME")
for i, input_ in enumerate(input_seq):
    # should get the same values here
    print(rn(input_seq[0], i)[0, 0, 0])

print("=" * 100)
l1_loss = nn.L1Loss()
pred = 0
for i, input_ in enumerate(input_seq):
    pred += torch.sum(rn(input_, i))
out = l1_loss(10, pred)
out.backward()
print("=" * 100)
    
print("THESE VALUES SHOULD BE DIFFERENT")
for i, input_ in enumerate(input_seq):
    # should get different values here
    print(rn(input_seq[0], i)[0, 0, 0])
```
